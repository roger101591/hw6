```python
from citipy import citipy
import matplotlib.pyplot as plt
import pandas as pd
import openweathermapy.core as owm
import random
```

```python
city = citipy.nearest_city(6,1)
city = city.city_name
city
```

```python
#get list of random cities
cities = []
while len(cities) < 500:
    latitude = random.randint(0,90)
    longitude = random.randint(0,180)
    city = citipy.nearest_city(latitude,longitude)
    city = city.city_name
    if city not in cities:
        cities.append(city)

citiesdf = pd.DataFrame(cities)
citiesdf
```

```python
#unique city check
city_list = citiesdf[0].value_counts()
city_list
```
